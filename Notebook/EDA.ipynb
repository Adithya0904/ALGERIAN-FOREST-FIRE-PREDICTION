```python
df_copy=pd.read_csv('Algerian_forest_fires_updated.csv')
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),'not fire','fire' )
```

```python
df_copy.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_copy.head()
```

```python
dfreg0=df_copy[df_copy['Region']==0]
sns.countplot(x=dfreg0['month'],hue=dfreg0['Classes'])
plt.xlabel('Month')
plt.ylabel('Fire vs no Fire')
plt.title('Fire Analysis in Bejaia Region')
```

```python
dfreg1=df_copy[df_copy['Region']==1]
sns.countplot(x=dfreg1['month'],hue=dfreg1['Classes'])
plt.xlabel('Month')
plt.ylabel('Fire vs no Fire')
plt.title('Fire Analysis in Sidi-Bel Abbes Region')
```
